# Arguments for Histogram 

```python

range[1000,5000] # autoscaling is based on the specified bin range instead of the range of x

bins  = 6 # set the bin size 

rwidth = 0.85 # gap between the bins 

color  = 'r' # color 

histtype{'bar', 'barstacked', 'step', 'stepfield'} # the type of the histogram to draw

orientation{'horizontal', 'vertical'} # type of view ; 

align{'left', 'mid', 'right'} # controls how the histogram is plotted default : mid

```

```python
plt.figure(figsize=(8,6))

plt.style.use('tableau-colorblind10')
day_wise_cases = [1755,2411,2487,2573,3875,2680, 3561,3390,3320,3277,4213]

plt.hist(day_wise_cases, range=[1000,5000],density=False, bins=6, rwidth=.85,
         histtype='barstacked', label='day_wise_cases', color='grey')

plt.legend()
plt.xlabel('Number of Cases')
plt.ylabel('counts')

plt.title('Day-Wise-Count')


plt.show()

# only 4 bars appeared because in first bin and sixt bin there are no data values so its empty. 
```
