# Subplot:
1. plt.subplots(nrows, ncols) creates a grid of plots.

2. axs is a 2D array of axes if nrows and ncols > 1.

3. Use axs[i, j] to access the subplot at row i and column j.

4. tight_layout() helps prevent label overlap.



```python
x = np.array([0,1,2,3,4])
y = np.array([3,10,12,14,4])

plt.subplot(1,3,1) # .subplot(1 row, 3 col, no of plots)
plt.plot(x,y)

x1 = np.array([0,1,2,3,4])
y1 = np.array([3,13,14,15,16])

plt.subplot(1,3,2) 
plt.plot(x1,y1)

x2 = np.array([0,1,2,3,4])
y2 = np.array([3,13,14,15,16])

plt.subplot(1,3,3) 
plt.plot(x2,y2)

```

```python
x = np.array([0,1,2,3,4])
y = np.array([3,10,12,14,4])

plt.subplot(3,1,1) # .subplot(3 row, 1 col, nth of plots)
plt.plot(x,y)

x1 = np.array([0,1,2,3,4])
y1 = np.array([3,13,14,15,16])

plt.subplot(3,1,2) 
plt.plot(x1,y1)

x2 = np.array([0,1,2,3,4])
y2 = np.array([3,13,14,15,16])

plt.subplot(3,1,3) 
plt.plot(x2,y2)
```

# 1 Line Plot

plot([x], y, [fmt], *, data=None, **kwargs)

plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

```python
# plot([x], y, [fmt], *, data=None, **kwargs)
# plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

# [x] is not mandatory by default given value will be taken as y axis.

# kwargs -> keyword arguments(optional)

# mec -> marker edge color 

fmt = '[marker][line][color]'

plt.plot([1,2,3,4,15], 'or:',[2,3,5,4,10], '*c--',mec='black', mfc='blue',ms=20, alpha=0.6)  # by default gives line plot 

# by default value given is considered as y axis (gets plotted with y axis)
```

# 2. x and y label

```python
plt.plot([1,2,3,4,15], 'or:',[2,3,5,4,10], '*c--',mec='black', mfc='blue',ms=20, alpha=0.9)  # by default gives line plot 

plt.xlabel("x label", fontdict={'size': 20, 'color': 'gray'}, labelpad=20, loc='right')
plt.show()

```

```python
plt.plot([1,2,3,4,15], 'or:',[2,3,5,4,10], '*c--',mec='black', mfc='blue',ms=20, alpha=0.9)  # by default gives line plot 

plt.ylabel("y label", fontdict={'size': 20, 'color': 'red'}, labelpad=20, loc='top')
plt.show()
```

# 3. Title 

```python
plt.plot([1,2,3,4,15], 'or:',[2,3,5,4,10], '*c--',mec='black', mfc='blue',ms=20, alpha=0.9)  # by default gives line plot 

plt.title('sample', fontdict={'size' : 15, 'color': 'green'}, pad=20, loc='left')
plt.ylabel("y label", fontdict={'size': 20, 'color': 'red'}, labelpad=20, loc='top')
plt.show()
```

# 4 Grid 

```python
plt.plot([1,2,3,4,15], 'or:',[2,3,5,4,10], '*c--',mec='black', mfc='blue',ms=20, alpha=0.9)  # by default gives line plot 
plt.title('sample', fontdict={'size' : 15, 'color': 'green'}, pad=20, loc='left')

plt.minorticks_on()
plt.grid(which= 'minor', color='black', axis='both')

plt.grid(visible = False,which='major', color='red')
plt.show()
```

# 5 xlim & ylim

```python
plt.plot([1,2,3,4,5,6],[2,3,4,5,6,7])

plt.xlim(0,7)
plt.ylim(1,8)
```

```python
plt.plot([1,2,3,4,5,6],[2,3,4,5,6,7], linestyle='dotted', linewidth=5)

plt.xlim(0,7)
plt.ylim(1,8)
plt.xticks(np.arange(4), ['arun','bal','car','dog'] , rotation=90)
```
