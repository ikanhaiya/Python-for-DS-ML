```python
#Get the avg total_bill by only male
a = tip.loc[tip['sex']=='Male']

m = a['total_bill'].mean() 
m = a.mean(numeric_only=True)
m
```

```python
#Get the avg(mean,median) total_bill and tip given by smoker and non smoker

total_bill_mean_smoker = tip.loc[tip['smoker']=='Yes']['total_bill'].mean()

total_bill_mean_NonSmoker = tip[tip['smoker']=='No']['total_bill'].mean()
total_bill_mean_NonSmoker

tip_mean_smoker = tip.loc[tip['smoker']=='Yes']['tip'].mean()
tip_mean_smoker
tip_mean_NonSmoker = tip[tip['smoker']=='No'].loc[:,'tip'].mean()
tip_mean_NonSmoker

total_bill_median_smoker = tip.loc[tip['smoker']=='Yes']['total_bill'].median()
total_bill_median_smoker

total_bill_median_NonSmoker = tip.loc[tip['smoker']=='No']['total_bill'].median()
total_bill_median_NonSmoker

tip_median_smoker = tip[tip['smoker']=='Yes']['tip'].median()
tip_median_smoker
tip_median_NonSmoker = tip[tip['smoker']=='No']['tip'].median()
tip_median_NonSmoker
```

```python
# finding the total_bill mean of non smokers and where total_bill is greater than 12

a = tip[(tip['total_bill'] >12) & (tip['smoker']=='No')]['total_bill'].mean()
a
```

```python
# what is most frequent time of vist

a = tip['time'].mode()
a
# Dinner time is bussiest.
```

```python
#Which day customers are giving higest tips

# 1. To calculate which day customers are giving highest tips 
# 2. We need to group tips daywise then calculate mean of every day.
# 3. On whichever day mean is maxm that day customer gives higest tip.
# 3.2551315789473683

a = tip.groupby(['day'])['tip'].mean().idxmax()
a

# idxmax is used to find the index of the maximum element in series. 

# using for loop 
a = tip.groupby('day')['tip'].mean()
mx = max(a)
idx = a.index
idx
for i in idx:
    if a[i] == mx:
        print(i)

```

```python
#get the avg(median) tip and total_bill amount by smoker(yes) on sunday

# 1. median of total_bill and tip by smokers on sunday.
# 2. 2 filters smokers == yes and day==sunday.

a  = tip[(tip['smoker']=='Yes') & (tip['day']=='Sun')][['tip','total_bill']].median()
a
```

```python
#get the (mean,median,min,max) tip and total_bill amount by smoker(yes) on friday with time(Lunch)

# 3 filters smoker==yes and day ==friday and time ==lunch
# mean of tip and total_bill 

# .agg(['mean','median','min','max'])

mean = tip.loc[(tip['smoker']=='Yes') & (tip['day']=='Fri') & (tip['time']=='Lunch')][['tip','total_bill']].mean()
mean

median = tip[(tip['smoker']=='Yes') & (tip['day']=='Fri') & (tip['time']=='Lunch')][['tip','total_bill']].median()
median

min = tip.loc[(tip['smoker']=='Yes') & (tip['day']=='Fri') & (tip['time']=='Lunch')][['tip','total_bill']].min()
min

max = tip.loc[(tip['smoker']=='Yes') & (tip['day']=='Fri') & (tip['time']=='Lunch')][['tip','total_bill']].max()
max

# Using agg function to calculate all this using one lines only

a = tip.loc[(tip['smoker']=='Yes') & (tip['day']=='Fri') & (tip['time']=='Lunch')][['tip','total_bill']].agg(['mean','median','min','max'])
a



#1 What is min bill by a smoker on sunday

a = tip[(tip['smoker']=='Yes') & (tip['day']=='Sun')]['total_bill'].min()
a

#2 what is max tip by a non smoker on thursday

a = tip[(tip['smoker']=='No') & (tip['day']=='Thur')]['tip'].max()
a

#3 what is total_bill made on sat by non smokers

a = tip[(tip['smoker']=='No') & (tip['day']=='Sat')]['total_bill'].sum()
a

```

```python
#create a new column total_bill amount category(low, moderate, high)

def categorize_bill(amount):
    if amount < 10:
        return 'Low'
    elif amount < 30:
        return 'Moderate'
    else:
        return 'High'

tip['bill_category'] = tip['total_bill'].apply(categorize_bill)
tip
```
