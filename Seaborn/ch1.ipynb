```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("tips")
data.head()
```

1. Relplot

```python
sns.relplot(data = data, x = "total_bill", y = "tip", kind='scatter', col = 'sex', row = 'smoker', hue = 'time', size = 'day', style = 'size') 
```

# 3. Scatterplot

```python
  sns.scatterplot(data = data, x = "total_bill", y = "tip")
```

```python
sns.lineplot(data = data, x = "total_bill", y = "tip", hue = 'time') 
```

# 4. Distplot

```python
sns.displot(data['tip'], kde = True)

# Kernel Density Estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
# It smooths the data points into a continuous curve, allowing for a better understanding of the distribution of the data.
# It is particularly useful for visualizing the distribution of a dataset, especially when the data is not normally distributed.
# KDE is often used in data analysis and visualization to identify patterns, outliers, and the overall shape of the data distribution.
```

# 5. Barplot

```python
sns.barplot(data = data, x = "day", y = "tip", estimator = np.min)
```

# 3. Countplot 

```python
sns.countplot(data = data, x = 'sex', hue = 'time')
```

# 7. Boxenplot 

```python
sns.boxenplot(data = data, x = 'sex', y = 'total_bill', hue = 'time')
```

# 8. Violin 

```python
sns.violinplot(data = data, x = 'sex', y = 'total_bill', hue = 'time')
```

# 9. Jointplot 

```python
 sns.jointplot(data = data, x = 'total_bill', y = 'tip')
```

```python
 sns.jointplot(data = data, x = 'total_bill', y = 'tip', kind = 'kde')
```
